```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from sklearn import tree
from sklearn import svm
```

```python
my_data = pd.read_csv("Titanic-my_dataset (1).csv")
```

```python
my_data
```

```python
my_data = my_data.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
my_data
```

```python
my_data.isnull().sum()
```

```python
median = my_data['Age'].median()
```

```python
my_data['Age']=my_data['Age'].fillna(median)
```

```python
max = my_data['Embarked'].value_counts().idxmax()
```

```python
my_data['Embarked']=my_data['Embarked'].fillna(max)
```

```python
my_data = pd.get_dummies(my_data,columns=['Embarked','Sex'])
my_data
```

```python
y = pd.my_dataFrame(my_data['Survived'], columns=['Survived'])
```

```python
x = my_data.drop(columns=['Survived'])
```

```python
x.head()
```

```python
x.shape
```

```python
y.head()
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=69)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
y_train = np.ravel(y_train)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train_scaled, y_train)

pred = model.predict(x_test_scaled)
```

```python
print(f1_score(y_test, pred))
```

```python
mymodel = svm.SVC(C=1.0, kernel='rbf')
mymodel.fit(x_train_scaled, y_train)
```

```python
mymodel.score(x_test_scaled, y_test)
```

```python
tree_model = tree.DecisionTreeClassifier(criterion='gini')
tree_model.fit(x_train, y_train)
```

```python
tree_model.score(x_test, y_test)
```
